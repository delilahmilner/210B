```python
import numpy as np
import matplotlib as plt
from datascience import*
%matplotlib inline 
plt.style.use('fast')
from scipy import special as special
```

```python
# number 2 part c graphing part

h = 1
Bo = (1.99/8065.54) / 27.2114
j2 = 2
E2 = Bo*j2*(j2+1)
j0 = 0
E0 = Bo*j0*(j0+1)
He = 4*1822.89/2
N2 = 28*1822.89/2
u = ((He*N2)/(He+N2)) * 27.211
# if in eV u*27.11
w = (E2-E0)/h
Umn = 10**-5
ao = 1
Beta = 1 /((4*ao)**2)  
#
A = (u * w**2)/ (4*Beta)
```

```python
A
```

```python

Energy = np.linspace(0.0000001,0.01,100)
cross_section = []
for E in Energy: 
    c = (u*(np.pi**2))/(4*(Beta**2)*(h**2)) *  Umn**2 * (1/E) * np.exp((-(1 * A) / E))
    cross_section.append(c)
```

```python
Table().with_columns("Energy",Energy,"Cross Section",cross_section).plot(0)
```

```python
# number 2 part f
Kb = 3.167e-6
```

```python
Temp = np.linspace(1,1000,500)
rate_constant = []
for T in Temp: 
    z = np.sqrt(((u*(w**2)))/((Beta*Kb*T)))
    First_term = ((2*np.pi)**(3/2)) / (np.sqrt(Kb*T))  
    Second_term = (np.sqrt(u)) / (4*((Beta**2)*(h**2)))
    Third_term = Umn**2
    Fourth_term = np.sqrt ( ((u) * w**2) / ((Beta) * Kb * T) )
    Fifth_term = special.kv(1,z)
    k_T = First_term*Second_term*Third_term*Fourth_term*Fifth_term
    rate_constant.append(k_T)
```

```python
Table().with_columns("Temperature",Temp,"Rate Constant",rate_constant).plot(0)
```
